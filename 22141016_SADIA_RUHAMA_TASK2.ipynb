```python
import sys
from collections import defaultdict

def find_eulerian_path(graph_adj):

    # 1. Calculate in-degrees and out-degrees for all nodes
    in_degree = defaultdict(int)
    out_degree = defaultdict(int)
    all_nodes = set() # Keep track of all nodes encountered

    # Populate degrees and all_nodes set
    for u, neighbors in graph_adj.items():
        all_nodes.add(u)
        out_degree[u] += len(neighbors)
        for v in neighbors:
            all_nodes.add(v)
            in_degree[v] += 1

    # 2. Find the starting node for the Eulerian path

    start_node = None
    for node in all_nodes:
        if out_degree[node] - in_degree[node] == 1:
            start_node = node
            break


    if start_node is None:
        for node in all_nodes:
            if out_degree[node] > 0:
                start_node = node
                break

        if start_node is None:
            return []

    # 3. Implement Hierholzer's algorithm

    modified_graph = {node: list(neighbors) for node, neighbors in graph_adj.items()}
    for node in modified_graph:
        modified_graph[node].reverse()

    current_path_segment_stack = [start_node]
    euler_path_reversed = []

    current_node = start_node

    while current_path_segment_stack:

        if current_node in modified_graph and modified_graph[current_node]:

            current_path_segment_stack.append(current_node)
            current_node = modified_graph[current_node].pop()
        else:

            euler_path_reversed.append(current_node)
            current_node = current_path_segment_stack.pop()


    return euler_path_reversed[::-1]

def parse_graph_from_file(filename):

    graph_adj = defaultdict(list)
    try:
        with open(filename, 'r') as f:
            for line in f:
                line = line.strip()
                if not line:
                    continue

                parts = line.split(' -> ')
                u = int(parts[0])

                if len(parts) > 1:
                    neighbors_str = parts[1]
                    neighbors = list(map(int, neighbors_str.split(',')))
                    graph_adj[u].extend(neighbors)
    except FileNotFoundError:
        raise FileNotFoundError(f"The input file '{filename}' was not found.")
    except Exception as e:
        raise Exception(f"Error parsing file '{filename}': {e}")

    return graph_adj

if __name__ == "__main__":

    input_filename = "/content/rosalind_ba3g.txt"

    try:

        graph = parse_graph_from_file(input_filename)

        # Find the Eulerian path
        path = find_eulerian_path(graph)


        if path:
            print("->".join(map(str, path)))
        else:
            print("No Eulerian path found or graph is empty/invalid.")

    except FileNotFoundError as e:
        print(f"Error: {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
```
